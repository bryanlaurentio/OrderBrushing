```python
import pandas as pd
```

```python
#read dataframe
df = pd.read_csv('order_brush_order.csv')
df.info()
```

```python
df.head()
```

```python
#ubah event time jadi index buat pakai pd.Grouper 
df = df.set_index(pd.DatetimeIndex(df['event_time']))
```

```python
df = df.drop(columns = 'event_time')
```

```python
df
```

```python
df_group = df.groupby(['shopid','userid',pd.Grouper(freq='60min')]).count()
```

```python
df_group
```

```python
df_group = df_group.rename(columns = {'orderid':'n_order'})
```

```python
df_group
```

```python
#cari n_order yang lebih dari sama dengan 3
df_suspect = df_group[df_group['n_order'] > 2]
```

```python
df_suspect
```

```python
df_suspect = df_suspect.reset_index()
```

```python
df_suspect
```

```python
print('Jumlah toko yang dicurigai {}'.format(len(df_suspect['shopid'].unique())))
print('Jumlah user yang order lebih dari 3 dalam 1 jam {}'.format(len(df_suspect['userid'].unique())))
```
